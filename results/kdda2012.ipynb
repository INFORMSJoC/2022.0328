```python
import matplotlib.pyplot as plt
```

```python
cores = ['serial', '8', '16', '24', '36', '40', '48', '56', '64', '72', '80']
```

```python
condat_1 = [201.567, 38.679, 26.658, 21.371, 18.596, 16.678, 15.271, 14.933, 14.594, 13.838, 14.454]
michelot_1 = [1950, 262.215, 127.998, 91.890, 71.800, 64.834, 58.330, 53.871, 51.130, 49.323, 56.009 ]
ss_1 = [3.685, 1.54, 1.230, 1.155, 1.117, 1.119, 1.088, 1.102, 1.062, 1.079, 1.0893]
sps_1 = [3.685, 1.48, 1.186, 1.117, 1.061, 1.072, 0.983, 0.991, 1.006, 0.973, 0.976]
condat_1_rate = [0]*11
michelot_1_rate = [0]*11
ss_1_rate = [0]*11
sps_1_rate = [0]*11
for i in range(11):
    condat_1_rate[i] = condat_1[0]/condat_1[i]
    michelot_1_rate[i] = michelot_1[0]/michelot_1[i]
    ss_1_rate[i] = ss_1[0]/ss_1[i]
    sps_1_rate[i] = sps_1[0]/sps_1[i]
```

```python
plt.plot(cores[1:11], michelot_1_rate[1:11], 'o-', label = "(P)Michelot", color="blue")
plt.plot(cores[1:11], condat_1_rate[1:11], 's-', label = "(P)Condat", color="red")
plt.plot(cores[1:11], sps_1_rate[1:11], '<-', label = "(P)Sort and (P)Scan", color="orange")
plt.plot(cores[1:11], ss_1_rate[1:11], '>-', label = "(P)Sort and Scan ", color="green")
#plt.ylabel("runtime/serial runtime")
plt.title("Serial runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 500
plt.savefig("kdd2012_1.png", format="png", dpi=resolution_value)
plt.show()
```

```python
condat_1_comp = [0]*10
michelot_1_comp = [0]*10
ss_1_comp = [0]*10
sps_1_comp = [0]*10
for i in range(10):
    condat_1_comp[i] = condat_1[0]/condat_1[i+1]
    michelot_1_comp[i] = condat_1[0]/michelot_1[i+1]
    ss_1_comp[i] = condat_1[0]/(ss_1[i+1]*1000)
    sps_1_comp[i] = condat_1[0]/(sps_1[i+1]*1000)
```

```python
plt.plot(cores[1:11], condat_1_comp, 's-', label = "(P)Condat", color="red")
plt.plot(cores[1:11], michelot_1_comp, 'o-', label = "(P)Michelot", color="blue")
plt.plot(cores[1:11], sps_1_comp, '<-', label = "(P)Sort and (P)Scan", color="orange")
plt.plot(cores[1:11], ss_1_comp, '>-', label = "(P)Sort and Scan ", color="green")
#plt.ylabel("runtime/serial runtime")
plt.title("Condat runtime/parallel runtime v.s. number of cores")
plt.legend(loc = "upper left")
resolution_value = 500
plt.savefig("kdd2012_1_comp.png", format="png", dpi=resolution_value)
plt.show()
```
