```python
import matplotlib.pyplot as plt
```

```python
cores = [1, 8, 16, 24, 32, 40, 48, 56, 64, 72, 80]
```

```python
condat = [3.227, 1.336, 1.234, 1.211, 1.205, 1.198, 1.216, 1.198, 1.205, 1.215, 1.216]
michelot = [6.848, 1.788, 1.480, 1.405, 1.340, 1.311, 1.288, 1.285, 1.273, 1.271, 1.277]
ss = [13.722, 5.053, 4.053, 3.843, 3.649, 3.563, 3.564, 3.541, 3.513, 3.546, 3.601]
sps = [13.722, 4.923, 3.873, 3.670, 3.548, 3.441, 3.437, 3.387, 3.391, 3.419, 3.427]
condat_rate = [0]*11
michelot_rate = [0]*11
ss_rate = [0]*11
sps_rate = [0]*11
for i in range(11):
    condat_rate[i] = condat[0]/condat[i]
    michelot_rate[i] = michelot[0]/michelot[i]
    ss_rate[i] = ss[0]/ss[i]
    sps_rate[i] = sps[0]/sps[i]
```

```python
plt.plot(cores[1:11], michelot_rate[1:11], 'o-', label = "(P)Michelot", color = "blue")
plt.plot(cores[1:11], sps_rate[1:11], '<-', label = "(P)Sort and (P)Scan", color = "orange")
plt.plot(cores[1:11], ss_rate[1:11], '>-', label = "(P)Sort and Scan ", color = "green")
plt.plot(cores[1:11], condat_rate[1:11], 's-', label = "(P)Condat", color = "red")
plt.ylabel("relative speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Serial runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("pptotal.png", format="png", dpi=resolution_value)
plt.show()
```

```python
condat_comp = [0]*10
michelot_comp = [0]*10
ss_comp = [0]*10
sps_comp = [0]*10
for i in range(10):
    condat_comp[i] = condat[0]/condat[i+1]
    michelot_comp[i] = condat[0]/michelot[i+1]
    ss_comp[i] = condat[0]/ss[i+1]
    sps_comp[i] = condat[0]/sps[i+1]
```
