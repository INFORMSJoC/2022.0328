```python
import matplotlib.pyplot as plt
```

```python
b = ["1", "10", "100", "1000", "10000", "100000", "1000000"]
```

```python
condat_s = [272.668, 235.772, 264.174, 258.757, 310.096, 476.168, 1150]
condat_p = [18.113, 19.985, 36.371, 76.343, 99.514, 442.031, 1097]
condat_s_100 = [235.787, 235.590, 232.730, 238.978, 240.327, 300.686, 309.617]
condat_p_100 = [18.458, 17.702, 18.281, 19.759, 22.867, 35.100, 94.282]
michelot_s = [3763, 3711, 3694, 3703, 3726, 3740, 3882]
michelot_p = [182.245, 181.881, 181.925, 190.866, 247.279, 655.864, 2423]
michelot_s_100 = [3728, 3715, 3764, 3714, 3715, 3689, 3712]
michelot_p_100 = [185.529, 184.649, 181.974, 182.033, 181.262, 190.210, 251.157]
sortscan_s = [10286, 10336, 10285, 10340, 10307, 10341, 10473]
sortscan_p = [2465, 2472, 2463, 2502, 2493, 2456, 2586, 2524]
sortPscan_p = [2346, 2377, 2372, 2371, 2361, 2517, 2415]
sortscan_s_100 = [10381, 10285, 10352, 10257, 10282, 10291, 10292]
sortscan_p_100 = [2451, 2486, 2466, 2458, 2458, 2463, 2480]
sortPscan_p_100 = [2266, 2269, 2314, 2359, 2334, 2379, 2367]
```

```python
condat_rate_re = [0]*7
condat_rate_ab = [0]*7
michelot_rate_re = [0]*7
michelot_rate_ab = [0]*7
sortscan_rate_re = [0]*7
sortscan_rate_ab = [0]*7
sortPscan_rate_re = [0]*7
sortPscan_rate_ab = [0]*7
for i in range(7):
    condat_rate_re[i] = condat_s[i]/condat_p[i]
    condat_rate_ab[i] = condat_s[i]/condat_p[i]
    michelot_rate_re[i] = michelot_s[i]/michelot_p[i]
    michelot_rate_ab[i] = condat_s[i]/michelot_p[i]
    sortscan_rate_re[i] = sortscan_s[i]/sortscan_p[i]
    sortscan_rate_ab[i] = condat_s[i]/sortscan_p[i]
    sortPscan_rate_re[i] = sortscan_s[i]/sortPscan_p[i]
    sortPscan_rate_ab[i] = condat_s[i]/sortPscan_p[i]
```

```python
plt.plot(b, michelot_rate_re, 'o-', label = "(P)Michelot", color = "blue")
plt.plot(b, condat_rate_re, 's-', label = "(P)Condat", color = "red")
plt.plot(b, sortPscan_rate_re, '<-', label = "(P)Sort and (P)Scan", color = "orange")
plt.plot(b, sortscan_rate_re, '>-', label = "(P)Sort and Scan", color = "green")
plt.ylabel("relative speedup")
plt.xlabel("scaling factor b")
plt.xticks(b)
#plt.title("Serial runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("robust_re_1.png", format="png", dpi=resolution_value)
plt.show()
```

```python
plt.plot(b, condat_rate_ab, 's-', label = "(P)Condat", color = "red")
plt.plot(b, michelot_rate_ab, 'o-', label = "(P)Michelot", color = "blue")
plt.plot(b, [1]*7, '--', label = "(S)Condat", color = "grey")
plt.plot(b, sortPscan_rate_ab, '<-', label = "(P)Sort and (P)Scan", color = "orange")
plt.plot(b, sortscan_rate_ab, '>-', label = "(P)Sort and Scan", color = "green")
plt.ylabel("absolute speedup")
plt.xlabel("scaling factor b")
plt.xticks(b)
#plt.title("Serial runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("robust_ab_1.png", format="png", dpi=resolution_value)
plt.show()
```
