```python
import matplotlib.pyplot as plt
```

```python
cores = [1, 8, 16, 24, 32, 40, 48, 56, 64, 72, 80]
```

```python
condat = [886.668, 468.221, 256.072, 221.507, 205.050, 193.569, 198.630, 162.498, 144.340, 149.794, 137.418]
michelot = [6491, 1395, 886.491, 807.825, 757.192, 652.471, 680.175, 585.993, 538.623, 443.252, 445.853]
ss = [278.963, 127.137, 123.372, 54.417, 51.668, 52.878, 59.342, 58.109, 59.984, 55.325, 57.542]
condat_rate = [0]*11
michelot_rate = [0]*11
ss_rate = [0]*11
for i in range(11):
    condat_rate[i] = condat[0]/condat[i]
    michelot_rate[i] = michelot[0]/michelot[i]
    ss_rate[i] = ss[0]/ss[i]
```

```python
plt.plot(cores[1:11], michelot_rate[1:11], 's-', label = "(P)Michelot", color = "blue")
plt.plot(cores[1:11], condat_rate[1:11], 'o-', label = "(P)Condat", color = "red")
plt.plot(cores[1:11], ss_rate[1:11], '>-', label = "(P)Sort and Scan ", color = "green")
plt.ylabel("relative speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Serial runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("wl1ball.png", format="png", dpi=resolution_value)
plt.show()
```

```python
condat_comp = [0]*10
michelot_comp = [0]*10
ss_comp = [0]*10
for i in range(10):
    condat_comp[i] = condat[0]/condat[i+1]
    michelot_comp[i] = condat[0]/michelot[i+1]
    ss_comp[i] = condat[0]/(ss[i+1]*1000)
```

```python
plt.plot(cores[1:11], condat_comp, 'o-', label = "(P)Condat", color = "red")
plt.plot(cores[1:11], michelot_comp, 's-', label = "(P)Michelot", color = "blue")
plt.plot(cores[1:11], [1]*10, '--', label = "(S)Condat", color = "grey")
plt.plot(cores[1:11], ss_comp, '>-', label = "(P)Sort and Scan ", color = "green")
plt.ylabel("absolute speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Condat runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("wl1ball_comp.png", format="png", dpi=resolution_value)
plt.show()
```
