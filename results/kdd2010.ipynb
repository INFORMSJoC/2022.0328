```python
import matplotlib.pyplot as plt
```

```python
cores = ["serial", "8", "16", "24", "32", "40"]
```

```python
condat_1 = [8.142, 5.555, 3.618, 2.984, 2.241, 2.589]
michelot_1 = [60.852, 26.012, 20.612, 15.580, 13.520, 16.533]
ss_1 = [99.082, 69.29, 50.415, 43.668, 42.547, 46.848]
sps_1 = [99.082, 63.036, 45.926, 41.603, 40.283, 40.778]
condat_1_rate = [0]*6
michelot_1_rate = [0]*6
ss_1_rate = [0]*6
sps_1_rate = [0]*6
for i in range(6):
    condat_1_rate[i] = condat_1[0]/condat_1[i]
    michelot_1_rate[i] = michelot_1[0]/michelot_1[i]
    ss_1_rate[i] = ss_1[0]/ss_1[i]
    sps_1_rate[i] = sps_1[0]/sps_1[i]
```

```python
plt.plot([8, 16, 24, 32, 40], michelot_1_rate[1:11], 'o-', label = "(P)Michelot", color = "blue")
plt.plot([8, 16, 24, 32, 40], condat_1_rate[1:11], 's-', label = "(P)Condat", color = "red")
plt.plot([8, 16, 24, 32, 40], sps_1_rate[1:11], '<-', label = "(P)Sort and (P)Scan", color = "orange")
plt.plot([8, 16, 24, 32, 40], ss_1_rate[1:11], '>-', label = "(P)Sort and Scan", color = "green")
plt.ylabel("relative speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40])
#plt.title("Serial runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("kdd2010_1.png", format="png", dpi=resolution_value)
plt.show()
```

```python
condat_1_comp = [0]*5
michelot_1_comp = [0]*5
ss_1_comp = [0]*5
sps_1_comp = [0]*5
for i in range(5):
    condat_1_comp[i] = condat_1[0]/condat_1[i+1]
    michelot_1_comp[i] = condat_1[0]/michelot_1[i+1]
    ss_1_comp[i] = condat_1[0]/ss_1[i+1]
    sps_1_comp[i] = condat_1[0]/sps_1[i+1]
```
