```python
condat_7_comp = [0]*10
michelot_7_comp = [0]*10
ss_7_comp = [0]*10
sps_7_comp = [0]*10
for i in range(10):
    condat_7_comp[i] = condat_7[0]/condat_7[i+1]
    michelot_7_comp[i] = condat_7[0]/michelot_7[i+1]
    ss_7_comp[i] = condat_7[0]/ss_7[i+1]
    sps_7_comp[i] = condat_7[0]/sps_7[i+1]
```
