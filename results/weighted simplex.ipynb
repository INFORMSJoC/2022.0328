```python
import matplotlib.pyplot as plt
```

```python
cores = [1, 8, 16, 24, 32, 40, 48, 56, 64, 72, 80]
```

```python
condat = [504.012, 87.081, 59.291, 46.517, 44.466, 42.564, 38.695, 38.939, 37.330, 35.747, 35.325]
michelot = [7274, 1203, 643.120, 473.265, 398.301, 318.840, 301.798, 301.455, 311.976, 256.150, 255.342]
ss = [303.231, 198.118, 154.392, 151.53, 129.429, 136.756, 127.543, 122.461, 123.482, 121.432, 122.349]
condat_rate = [0]*11
michelot_rate = [0]*11
ss_rate = [0]*11
for i in range(11):
    condat_rate[i] = condat[0]/condat[i]
    michelot_rate[i] = michelot[0]/michelot[i]
    ss_rate[i] = ss[0]/ss[i]
```

```python
plt.plot(cores[1:11], michelot_rate[1:11], 'o-', label = "(P)Weighted Michelot", color = "blue")
plt.plot(cores[1:11], condat_rate[1:11], 's-', label = "(P) Weighted Condat", color = "red")
plt.plot(cores[1:11], ss_rate[1:11], '>-', label = "(P) Weighted Sort and Scan ", color = "green")
plt.ylabel("relative speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Serial runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("wsimplex.png", format="png", dpi=resolution_value)
plt.show()
```

```python
condat_comp = [0]*10
michelot_comp = [0]*10
ss_comp = [0]*10
for i in range(10):
    condat_comp[i] = condat[0]/condat[i+1]
    michelot_comp[i] = condat[0]/michelot[i+1]
    ss_comp[i] = condat[0]/(ss[i+1]*1000)
```

```python
plt.plot(cores[1:11], condat_comp, 's-', label = "(P) Weighted Condat", color = "red")
plt.plot(cores[1:11], michelot_comp, 'o-', label = "(P)Weighted Michelot", color = "blue")
plt.plot(cores[1:11], [1]*10, '--', label = "(S)Condat", color = "grey")
plt.plot(cores[1:11], ss_comp, '>-', label = "(P) Weighted Sort and Scan ", color = "green")
plt.ylabel("absolute speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Condat runtime/parallel runtime v.s. number of cores")
plt.legend(loc = "upper left")
resolution_value = 300
plt.savefig("wsimplex_comp.png", format="png", dpi=resolution_value)
plt.show()
```
