```python
import matplotlib.pyplot as plt
```

```python
cores = [1, 8, 16, 24, 32, 40, 48, 56, 64, 72, 80]
```

```python
condat = [630.176, 68.084, 32.671, 27.533, 20.650, 18.136, 16.667, 15.100, 13.685, 13.159, 13.795]
condat_rate = [0]*11
for i in range(11):
    condat_rate[i] = condat[0]/condat[i]
```

```python
michelot = [3487, 534.458, 285.353, 193.471, 144.684, 116.952, 103.599, 98.175, 87.847, 74.992, 77.237]
michelot_rate = [0]*11
for i in range(11):
    michelot_rate[i] = michelot[0]/michelot[i]
```

```python
ss = [9.496, 3.802, 2.735, 2.473, 2.339, 2.306, 2.287, 2.268, 2.240, 2.267, 2.255]
sps = [9.496, 3.638, 2.397, 2.214, 2.168, 2.115, 2.072, 2.044, 2.102, 2.087, 2.061]
ss_rate = [0]*11
sps_rate = [0]*11
for i in range(11):
    ss_rate[i] = ss[0]/ss[i]
    sps_rate[i] = sps[0]/sps[i]
```

```python
plt.plot([8, 16, 24, 32, 40, 48, 56, 64, 72, 80], condat_rate[1:11], 's-', label = "(P)Condat", color="red")
plt.plot([8, 16, 24, 32, 40, 48, 56, 64, 72, 80], michelot_rate[1:11], 'o-', label = "(P)Michelot", color = "blue")
plt.plot([8, 16, 24, 32, 40, 48, 56, 64, 72, 80], sps_rate[1:11], '<-', label = "(P)Sort and (P)Scan", color="orange")
plt.plot([8, 16, 24, 32, 40, 48, 56, 64, 72, 80], ss_rate[1:11], '>-', label = "(P)Sort and Scan", color = "green")
plt.ylabel("relative speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("l1ball.png", format="png", dpi=resolution_value)
plt.show()
```

```python
condat_comp = [0]*10
michelot_comp = [0]*10
ss_comp = [0]*10
sps_comp = [0]*10
for i in range(10):
    condat_comp[i] = condat[0]/condat[i+1]
    michelot_comp[i] = condat[0]/michelot[i+1]
    ss_comp[i] = condat[0]/(ss[i+1]*(10**3))
    sps_comp[i] = condat[0]/(sps[i+1]*(10**3))
```

```python
plt.plot(cores[1:11], condat_comp, 's-', label = "(P)Condat", color="red")
plt.plot(cores[1:11], michelot_comp, 'o-', label = "(P)Michelot", color = "blue")
plt.plot(cores[1:11], [1]*10, '--', label = "(S)Condat", color = "grey")
plt.plot(cores[1:11], sps_comp, '<-', label = "(P)Sort and (P)Scan", color="orange")
plt.plot(cores[1:11], ss_comp, '>-', label = "(P)Sort and Scan", color = "green")
plt.ylabel("absolute speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Condat runtime/parallel runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("l1ball_comp.png", format="png", dpi=resolution_value)
plt.show()
```
