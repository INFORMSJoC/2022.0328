```python
condat_norm_rate = [0]*11
condat_unif_rate = [0]*11
condat_small_norm_rate = [0]*11
for i in range(11):
    condat_norm_rate[i] = condat_res_norm[0]/condat_res_norm[i]
    condat_unif_rate[i] = condat_res_uniform[0]/condat_res_uniform[i]
    condat_small_norm_rate[i] = condat_res_small_norm[0]/condat_res_small_norm[i]
```

```python
plt.plot(cores[1:11], condat_norm_rate[1:11], 'o-', label = r'$N(0,1)$', color = "blue")
plt.plot(cores[1:11], condat_unif_rate[1:11], '^-', label = r'$U[0,1]$', color = "green")
plt.plot(cores[1:11], condat_small_norm_rate[1:11], 's-', label = r'$N(0,0.001)$', color = "red")
plt.ylabel("relative speedup")
plt.xlabel("number of cores")
plt.xticks([8, 16, 24, 32, 40, 48, 56, 64, 72, 80])
#plt.title("Condat's method: serial runtime/runtime v.s. number of cores")
plt.legend()
resolution_value = 300
plt.savefig("simplex_condat.png", format="png", dpi=resolution_value)
plt.show()
```

```python
condat_norm_comp = [0]*10
michelot_norm_comp = [0]*10
ss_norm_comp = [0]*10
sps_norm_comp = [0]*10
for i in range(10):
    condat_norm_comp[i] = condat_res_norm[0]/condat_res_norm[i+1]
    michelot_norm_comp[i] = condat_res_norm[0]/michelot_res_norm[i+1]
    ss_norm_comp[i] = condat_res_norm[0]/(ss_res_norm[i+1]*(10**(3)))
    sps_norm_comp[i] = condat_res_norm[0]/(sps_res_norm[i+1]*(10**(3)))
```
